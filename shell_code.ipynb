```python
relatives_timeout = lambda x: x.timeout - timezone.now()  if x.timeout else 'Just now'

print(acc1.is_active_with_timeout())
print(acc2.is_active_with_timeout()) # When False is deactivated and timeout is None, it should show 'Just now'
print(acc3.is_active_with_timeout())
print('------------------------------')
print(relatives_timeout(acc1))
print(relatives_timeout(acc2)) # Should show approximately 1 minute
print(relatives_timeout(acc3))

# Active | Timeout 
# True   | None    => Active status 
# False  | None    => Deactivated status, but just now. because his user is decreased just now
# True   | Future  => Deactivated status, and the remaining time until the account is reactivated. a while banned as timeout.
# False  | Future  => Deactivated status, but deactivated is not just now. and the remaining time until the account is reactivated. a while banned as timeout.
```
